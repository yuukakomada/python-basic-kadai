```python
class Human:
    def __init__(self,name,age):
      self.name = name
      self.age = age

    def check_adult(self):
      if self.age >= 20:
        print("大人です")
      else:
        print("大人ではありません")

human1 = Human("侍太郎",18)
human2 = Human("侍一郎",19)
human3 = Human("侍二郎",20)
human4 = Human("侍花子",25)
human5 = Human("侍舞子",26)

human_list  = [human1,human2,human3,human4,human5]

for human in human_list:
  human.check_adult()
```
